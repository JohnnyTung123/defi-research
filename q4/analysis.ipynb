### Interaction Analysis
Etherscan has a hard limit of 10,000 transactions that can be obtained. Therefore, I only perform the analysis based on the most recent 10,000 transactions.

#### Transactions

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the transactions data
transactions_path = './csv/Transactions.csv'
df = pd.read_csv(transactions_path, index_col='Hash')
print("Transactions Data Shape:", df.shape)
df
```

```python
# Check for missing values
print("Missing Values in Transactions Data:")
print(df.isnull().sum())
```

```python
# Group transactions by sender-receiver pair and count
pair_counts = df.groupby(['From', 'To']).size().reset_index(name='Count')
print("Transaction counts for each sender-receiver pair:")
pair_counts
```

```python
# Calculate the range (min and max) of the 'Value_Eth' column
value_in_column = 'Value_Eth'
value_in_min = df[value_in_column].min()
value_in_max = df[value_in_column].max()
print(f'Range of Value_Eth: Min = {value_in_min}, Max = {value_in_max}')
```

```python
# Calculate the range (min and max) of the 'TransactionCost_Eth' column
txn_fee_column = 'TransactionCost_Eth'
txn_fee_min = df[txn_fee_column].min()
txn_fee_max = df[txn_fee_column].max()
print(f'Range of TransactionCost_Eth: Min = {txn_fee_min}, Max = {txn_fee_max}')
```

```python
# Convert the Timestamp column to datetime
df['TimeStamp'] = pd.to_datetime(df['TimeStamp'])

# Round the timestamp to the nearest minute (using 'min' instead of 'T')
df['Minute'] = df['TimeStamp'].dt.floor('min')  # or df['TimeStamp'].dt.round('min') for rounding

# Count transactions per minute
transaction_counts = df.groupby('Minute').size()

# Display the results
print("Transactions - Number of transactions")
print(transaction_counts)
```

```python
# Extract hour of the day
df['Hour'] = df['TimeStamp'].dt.hour

# Analyze transaction counts by hour
hourly_transaction_counts = df.groupby('Hour').size()
print("Transaction counts by hour of the day:")
print(hourly_transaction_counts)
```

```python
# Plotting
plt.figure(figsize=(12, 4))
hourly_transaction_counts.plot(kind='bar', color='skyblue')

# Add labels and title
plt.title('Transaction Counts by Hour', fontsize=14)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Transaction Count', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()

# Show the plot
plt.show()
```

#### Internal Transactions

```python
# Load the internal transactions data
internal_transactions_path = './csv/internal_transactions.csv'
df_internal = pd.read_csv(internal_transactions_path, index_col='Hash')
print("Internal Transactions Data Shape:", df_internal.shape)
df_internal
```

```python
# Check for missing values in internal transactions
print("Missing Values in Internal Transactions Data:")
print(df_internal.isnull().sum())
```

```python
# Group internal transactions by sender-receiver pair and count
pair_counts_internal = df_internal.groupby(['From', 'To']).size().reset_index(name='Count')
pair_counts_internal = pair_counts_internal.sort_values(by='Count', ascending=False)
print("Transaction counts for each sender-receiver pair in internal transactions:")
pair_counts_internal
```

```python
# Convert the Timestamp column to datetime
df_internal['TimeStamp'] = pd.to_datetime(df_internal['TimeStamp'])

# Round down the timestamp to the nearest minute
df_internal['Minute'] = df_internal['TimeStamp'].dt.floor('min')

# Count the number of internal transactions per minute
transaction_counts_per_minute = df_internal.groupby('Minute').size()

print("Internal Transactions - Number of transactions")
print(transaction_counts_per_minute.head(10))
```

```python
# Extract hour of the day
df_internal['Hour'] = df_internal['TimeStamp'].dt.hour

# Analyze transaction counts by hour
hourly_transaction_counts = df_internal.groupby('Hour').size()
print("Internal transaction counts by hour of the day:")
print(hourly_transaction_counts)
```

```python
# Plotting
plt.figure(figsize=(12, 4))
hourly_transaction_counts.plot(kind='bar', color='skyblue')

# Add labels and title
plt.title('Internal Transaction Counts by Hour', fontsize=14)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Transaction Count', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()

# Show the plot
plt.show()
```

#### ERC20_Transfers

```python
# Load the ERC-20 token transfers data
ERC20_transfers_path = './csv/erc20_transfers.csv'
df_erc20 = pd.read_csv(ERC20_transfers_path, index_col='Hash')
print("ERC-20 Transfers Data Shape:", df_erc20.shape)
df_erc20
```

```python
# Check for missing values in ERC-20 token transfers
print("Missing Values in ERC-20 Token Transfers Data:")
print(df_erc20.isnull().sum())
```

```python
# Count the number of token transfers by token name
token_transfer_counts = df_erc20['TokenName'].value_counts() 
print("Token transfer counts by TokenName:")
print(token_transfer_counts)
```

```python
# Convert UnixTimestamp to datetime
df_erc20['Timestamp'] = pd.to_datetime(df_erc20['TimeStamp'])

# Round down the timestamp to the nearest minute
df_erc20['Minute'] = df_erc20['Timestamp'].dt.floor('min')

# Count the number of ERC20 transactions per minute
transaction_counts_per_minute = df_erc20.groupby('Minute').size()

print("ERC20 transactions - Number of transactions")
print(transaction_counts_per_minute.head(10))
```

```python
# Extract hour of the day
df_erc20['Hour'] = df_erc20['Timestamp'].dt.hour

# Analyze transaction counts by hour
hourly_transaction_counts = df_erc20.groupby('Hour').size()
print("Transaction counts by hour of the day:")
print(hourly_transaction_counts)
```

```python
# Plotting
plt.figure(figsize=(12, 4))
hourly_transaction_counts.plot(kind='bar', color='skyblue')

# Add labels and title
plt.title('ERC20 transactions by Hour', fontsize=14)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Transaction Count', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()

# Show the plot
plt.show()
```
